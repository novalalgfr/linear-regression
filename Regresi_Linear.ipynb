# **Import Library**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

# **Business Understanding**

```python
#Tujuan: Membuat model untuk memprediksi harga rumah berdasarkan ukuran rumah.
```

# **Data Understanding**

```python
# Membaca data dari file CSV
df = pd.read_csv('_Synthetic House pricing dataset - Sheet1.csv')
```

```python
# Explorasi data
plt.scatter(df['House Size (sqft)'], df['Price (USD)'])
plt.xlabel("Ukuran Rumah (House Size (sqft))")
plt.ylabel("Harga (Price (USD))")
plt.title("Scatter Plot of House Size vs Price")
plt.show()
```

# **Data Preparation**

```python
# Mengatasi outlier atau data yang hilang (jika ada) dan membagi data
X = df[['House Size (sqft)']]
y = df['Price (USD)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# **Modeling**

```python
# Membuat model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)
```

# **Evaluation**

```python
# Evaluasi performa model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R^2 Score:", r2)
```

```python
# Visualisasi hasil prediksi dengan garis regresi
plt.scatter(df['House Size (sqft)'], df['Price (USD)'], color='blue', label='Data')
plt.plot(df['House Size (sqft)'], model.predict(df[['House Size (sqft)']]), color='red', label='Regression Line')
plt.xlabel("Ukuran Rumah (House Size (sqft))")
plt.ylabel("Harga (Price (USD))")
plt.legend()
plt.title("Model Regresi Linear")
plt.show()
```

# **Deployment (Simulation)**

```python
# Menggunakan model untuk prediksi harga rumah dengan ukuran tertentu
def predict_house_price(size):
    return model.predict(np.array([[size]]))[0]

print("Prediksi harga untuk rumah berukuran 1800 sq meters:", predict_house_price(1800))
```
